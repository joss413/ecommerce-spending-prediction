```python
from sklearn.model_selection import train_test_split
```

```python
X=df[["Avg. Session Length","Time on App","Time on Website","Length of Membership"]]
y=df["Yearly Amount Spent"]
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
df.shape
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
model.coef_,model.intercept_
```

```python
cdf=pd.DataFrame(model.coef_,X.columns,columns=['Coef'])
cdf
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x=y_test, y=y_pred, alpha=0.6)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)  # Root Mean Squared Error
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("MSE penalizes large errors heavily (lower is better).")
print("Mean Squared Error (MSE):", mse)
print("RMSE same as MSE but in the original units (lower is better).")
print("Root Mean Squared Error (RMSE):", rmse)
print("MAE average of absolute errors (lower is better).")
print("Mean Absolute Error (MAE):", mae)
print("When we evaluate R² Score : 1 is perfect, 0 means model is no better than average, negative means very poor fit.")
print("R² Score:", r2)
```
